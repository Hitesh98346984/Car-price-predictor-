# Cleaning

```python
backup = car.copy()
```

we are solving the first problem of data and we are keeping those year which are numeric, isnumeric() function return true or false if the value is numeric.

```python
car = car[car['year'].str.isnumeric()]
```

Next we will solve the 2nd problem where we have to change year to integer

```python
car['year'] = car['year'].astype(int)
```

```python
car.info()
```

now we will solve 3rd problem which is that there is "ask for price" as a column value but we want only numeric values

```python
car['Price'].unique
```

```python
car=car[car['Price']!="Ask For Price"]
```

now we will remove commas from price values, we are going to do it by replace these commas by empty strings and then we are going to convert this string to numeric using astype function.

```python
car['Price']=car['Price'].str.replace(',','').astype(int)
```

now we have to solve 5th problem which is "kms_driven has kms with interger"

```python
car['kms_driven']
```

```python
car['kms_driven']=car['kms_driven'].str.split(' ').str.get(0).str.replace(',','')
```

now we have to remove "petrol" value form the kms_driven column.

```python
car=car[car['kms_driven'].str.isnumeric()]
```

```python
car['kms_driven'] = car['kms_driven'].astype(int)
```

removing nan from fuel_type

```python
car[car['fuel_type'].isna()]
```

```python
car = car[~car['fuel_type'].isna()]
```

Now we have to keep first 3 value of the name

```python
car['name']=car['name'].str.split(' ').str.slice(0,3).str.join(' ')
```

```python
car = car.reset_index(drop=True)
```

removing outliers

```python
car[car['Price']>6e6]
```

```python
car = car[car['Price']<6e6].reset_index(drop=True)
```

# Model
features are everything except price column

```python
X = car.drop(columns='Price')
y = car['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe = OneHotEncoder()
ohe.fit(X[['name', 'company', 'fuel_type']])
```

```python
column_trans = make_column_transformer((OneHotEncoder(categories=ohe.categories_), ['name', 'company', 'fuel_type']), remainder='passthrough')
```

```python
lr = LinearRegression()
```

```python
pipe=make_pipeline(column_trans, lr)
```

```python
pipe.fit(X_train, y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
r2_score(y_test, y_pred)
```

```python
scores=[]
for i in range(1000):
    X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=i)
    lr = LinearRegression()
    pipe=make_pipeline(column_trans,lr)
    pipe.fit(X_train, y_train)
    y_pred=pipe.predict(X_test)
    scores.append(r2_score(y_test, y_pred))
```

```python
scores[np.argmax(scores)]
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=np.argmax(scores))
lr = LinearRegression()
pipe=make_pipeline(column_trans,lr)
pipe.fit(X_train, y_train)
y_pred=pipe.predict(X_test)
scores.append(r2_score(y_test, y_pred))
r2_score(y_test, y_pred)
```
